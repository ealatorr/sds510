```python
#imported necessary packages
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import time
#Established the keyword I was searching for. I decided to go for “cats” to try something fun and
#there is usually a lot of cat content on the internet. I will also be using mastodon for this search.
KEYWORD = "cats"                       
URL = f"https://mastodon.social/tags/{KEYWORD}"
CSV_FILE = "mod4_essentials_results.csv"     
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)
#Next I created the scraper I initially was having some difficulty but then I decided to make it simple and
#in this case worked for my results since there was only 20 results.
print("[+] Loading page:", URL)
driver.get(URL)

time.sleep(10) 

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

posts = soup.find_all("article")

print(f"[+] Found {len(posts)} posts on page.")

# Mapping it all into the final CSV file, then Printing Complete to make sure it is functioning
with open(CSV_FILE, "a", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    if f.tell() == 0:
        writer.writerow(["keyword", "snippet", "url"])
    for post in posts:
        text = post.get_text(" ", strip=True)
        if KEYWORD.lower() in text.lower():
            snippet = text[:200]
            link_tag = post.find("a", href=True)
            if link_tag:
                href = link_tag["href"]
                url = href if href.startswith("http") else "https://mastodon.social" + href
            else:
                url = "N/A"

            writer.writerow([KEYWORD, snippet, url])

print(f"[+] Done. Results saved to {CSV_FILE}")

driver.quit()
```
